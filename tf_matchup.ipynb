```python
def closest(list1, list2, k):
    return(list1[np.abs(list1 - list2[k]).argmin()])
```

```python
import matplotlib.pyplot as plt 

for chromosome in chromosomes[:2]:

    tmp = filtered_output[filtered_output['Chr'] == chromosome]['Unnamed: 9']

    labels, counts = np.unique(tmp,return_counts=True)
    ticks = range(len(counts))

    plt.title('Distribution of modules for chromosome '+ chromosome + " in P503_5.ATAC | Number of samples: " + str(len(tmp)) + '\n', fontsize = 20 )
    plt.ylabel('Counts', fontsize = 20 )

    bars = plt.bar(ticks,counts, align='center', linewidth=1, edgecolor=['black']*len(ticks))
    plt.xticks(ticks, labels)
    plt.xticks(rotation = 45, fontsize= 12)
    plt.yticks(fontsize = 15)
    plt.rcParams["figure.figsize"] = [30,5]
    plt.show()
```
